> In credit portfolio analysis, we usually assume that two firms from the same industry have a higher correlation (intra-correlation) $\rho_I$ than two firms from two different industries (inter-correlation) $\rho_O$. We usually use the following model:
$$
X_i^{(k)} =
\sqrt{\rho_I - \rho_O} \, X_k
+ \sqrt{\rho_O} \, X_M
+ \sqrt{1 - \rho_I} \, \varepsilon_i
\tag{1}
$$
where $X_k$, $X_M$, and $\varepsilon_i$ are all standard normal random variables with mean zero and variance one.  
They are also independent from each other, and $k$ represents a specific industry ($k = 1, 2, \ldots, K$).





<font color="#1E90FF" size="4"><b>№1 (a)</b></font>


> Verify that intra- and inter-industry correlations are $\rho_I$ and $\rho_O$

Let the latent variable for firm $i$ in industry $k$ be  
$$
X_i^{(k)} = \underbrace{\sqrt{\rho_I - \rho_O}}_{a} X_k
            + \underbrace{\sqrt{\rho_O}}_{b} X_M
            + \underbrace{\sqrt{1 - \rho_I}}_{c} \varepsilon_i,
$$
where  
$X_k, X_M, \varepsilon_i \sim \mathcal N(0,1)$ are mutually independent,    
$a = \sqrt{\rho_I - \rho_O}$, $b = \sqrt{\rho_O}$, $c = \sqrt{1 - \rho_I}$,  
and the parameters satisfy $0 \le \rho_O \le \rho_I \le 1$.

**Mean**  
Since each of the factors has a mean of 0:
$$
\mathbb{E}[X_i^{(k)}] = 0.
$$
$$
\mathbb{E}[X_i^{(k)}]
= a \mathbb{E}[X_k] + b \mathbb{E}[X_M] + c \mathbb{E}[\varepsilon_i]
= 0.
$$
**Variance**  
Because all components are independent,
$$
\begin{aligned}
\operatorname{Var}(X_i^{(k)})
&= \operatorname{Var}(a X_k + b X_M + c \varepsilon_i) \\
&= a^2 \operatorname{Var}(X_k) + b^2 \operatorname{Var}(X_M) + c^2 \operatorname{Var}(\varepsilon_i) \\
&= a^2 + b^2 + c^2 \\
&= (\rho_I - \rho_O) + \rho_O + (1 - \rho_I) = 1.
\end{aligned}
$$
Hence, each $X_i^{(k)}$ has **mean 0** and **variance 1**.  
Intra-Industry Correlation (Same Industry $k$)
For two different firms $i \ne j$ in the same industry $k$:

$$
X_i^{(k)} = a X_k + b X_M + c \varepsilon_i, \qquad
X_j^{(k)} = a X_k + b X_M + c \varepsilon_j.
$$

Their **covariance** is

$$
\begin{aligned}
\operatorname{Cov}(X_i^{(k)}, X_j^{(k)})
&= \mathbb{E}[X_i^{(k)} X_j^{(k)}] \\
&= a^2 \mathbb{E}[X_k^2] + b^2 \mathbb{E}[X_M^2] \\
&= a^2 + b^2 \\
&= (\rho_I - \rho_O) + \rho_O = \rho_I.
\end{aligned}
$$

Since each variance = 1, the **correlation** is

$$
\mathrm{Corr}(X_i^{(k)}, X_j^{(k)}) = \rho_I.
$$
Now take
$$
X_i^{(k)} = a X_k + b X_M + c \varepsilon_i, \qquad
X_j^{(\ell)} = a X_\ell + b X_M + c \varepsilon_j,
$$
where $X_k \perp X_\ell$.   
The **covariance** is

$$
\begin{aligned}
\operatorname{Cov}(X_i^{(k)}, X_j^{(\ell)})
&= \mathbb{E}[(a X_k + b X_M + c \varepsilon_i)
              (a X_\ell + b X_M + c \varepsilon_j)] \\
&= b^2 \mathbb{E}[X_M^2] = b^2 = \rho_O.
\end{aligned}
$$

Thus the **correlation** between firms in *different* industries is

$$
\mathrm{Corr}(X_i^{(k)}, X_j^{(\ell)}) = \rho_O.
$$
Hence, the model produces the desired **intra-industry correlation $\rho_I$** and **inter-industry correlation $\rho_O$**.

> Suppose there are a total of 12 different industries. How many independent factors do we need to represent the correlation matrix?

We use the factor model  
$$
X_i^{(k)} =
\sqrt{\rho_I - \rho_O} \, X_k
+ \sqrt{\rho_O} \, X_M
+ \sqrt{1 - \rho_I} \, \varepsilon_i,
$$
where:
- $X_M$ is the **market factor** common to all firms,  
- $X_k$ is the **industry factor** for industry $k$,  
- and $\varepsilon_i$ is the **idiosyncratic (firm-specific)** component.  
Each **industry** has its own specific random factor $X_k$, and all industries share one **global market factor** $X_M$.  
Therefore, if there are 12 industries:
$$
\text{Total number of independent factors} = 12 \text{ (industry)} + 1 \text{ (market)} = 13.
$$
The idiosyncratic terms $\varepsilon_i$ are not counted because they affect only individual firms and do not influence correlations between firms.

Special cases:
- If $\rho_I = \rho_O$: all industries have the same correlation: only the **market factor** matters.  
- If $\rho_O = 0$: no market correlation: only **12 industry factors** are needed.

>  Perform a PAA with a total of factors equal to 2, 3, 4, 5, and 6. Show the difference between the original correlation matrix and approximate correlation matrix, and draw your conclusion for the choice of number of independent factors in PAA.

We approximate $\Sigma$ using a **principal-axis approximation (PAA)** with $m$ common factors by keeping the first $m$ principal directions and then restoring unit diagonals:
$$
\Sigma_m^{\text{PAA}}
=\sum_{r=1}^{m}\lambda_r q_r q_r^\top
+\Big(I-\mathrm{diag}\!\big(\sum_{r=1}^{m}\lambda_r q_r q_r^\top\big)\Big),
$$
so that $\mathrm{diag}(\Sigma_m^{\text{PAA}})=1$ for all $m$.   
For the two-level (market + industry) structure, the eigenvalues are
$$
\lambda_1=2.3,\quad
\lambda_{2-5}=1.3,\quad
\lambda_{6-10}=0.5,
$$
and they sum to $\mathrm{tr}(\Sigma)=10$, as required for a $10\times10$ correlation matrix.    
The approximation error is defined as
$$
\Delta_m=\Sigma-\Sigma_m^{\text{PAA}}.
$$
Because the diagonal is forced to 1, the off-diagonal entries of $\Delta_m$ represent the contribution of omitted eigenpairs:
$$
(\Delta_m)_{ij}
=\sum_{r=m+1}^{10}\lambda_r\,q_r(i)\,q_r(j),
\qquad i\neq j.
$$
As $m$ increases, $\Delta_m$ shrinks — more co-movement structure is captured.  
| $m$ | Interpretation |
|:--:|:--|
| 2 | Captures the market mode ($\lambda_1$) and one industry contrast. Cross-industry correlations roughly match, but within-industry levels are still off. |
| 3–4 | Adds additional industry contrasts; within-industry correlations improve. |
| 5 | Market + all four independent industry modes → full market/industry co-movement. Only within-industry “difference” directions remain omitted. |
| 6 | Adds one within-industry contrast; aligns with the original two-level (market + five industry) model and reproduces all off-diagonal entries exactly. |   

A compact scalar proxy for the remaining co-movement is the **spectral residual energy**
$$
E_m=\sum_{r=m+1}^{10}\lambda_r^2.
$$
With $1.3^2=1.69$ and $0.5^2=0.25$, we get
| $m$ | Formula | $E_m$ |
|:--:|:--|:--:|
| 2 | $3\times1.3^2+5\times0.5^2$ | 6.32 |
| 3 | $2\times1.3^2+5\times0.5^2$ | 4.63 |

Smaller $E_m$ means less structure left out.

> Suppose that we have a one-period default swap over time interval [0, ∆t]. The premium P is paid at the beginning of the period, and the loss is paid at the time when default occurs, 1 − R, where R is the recovery rate. We assume that the hazard rate for the underlying credit is a constant h, and that we have a constant interest rate r over the time interval.  

<font color="#1E90FF" size="4"><b>№2 (а)</b></font>

>  Calculate the expected present value of loss leg for the one period default
swap;

Let $X$ denote the loss payment in a one-period default swap over $[0, \Delta t]$.  
If default occurs at time $\tau \le \Delta t$, the loss leg pays $(1 - R)$ discounted by $e^{-r\tau}$.  
With constant hazard rate $h$, the density of default time is  
$$
f_\tau(t) = h e^{-ht}, \quad t \ge 0.
$$
Hence, the expected present value is  
$$
\text{EPV}_{\text{loss}}
= (1 - R) \int_0^{\Delta t} e^{-rt} f_\tau(t)\,dt
= (1 - R) \int_0^{\Delta t} e^{-rt} h e^{-ht}\,dt.
$$
Evaluating,  
$$
\text{EPV}_{\text{loss}}
= (1 - R)\, h \int_0^{\Delta t} e^{-(r + h)t}\,dt
= (1 - R)\, \frac{h}{r + h}\, \big(1 - e^{-(r + h)\Delta t}\big).
$$

<font color="#1E90FF" size="4"><b>№2 (b)</b></font>

>  If default occurs at time $t$, and $0 \le t \le \Delta t$, then default protection buyer would receive a refund of $P(1 - \dfrac{t}{\Delta t})$ at the time of default from the protection seller. Calculate the expected present value of this refund.

Refund at default time $t\in[0,\Delta t]$ equals $P\!\left(1-\dfrac{t}{\Delta t}\right)$.  
Default time $\tau$ has constant hazard $h$ (so $f(t)=h e^{-ht}$), discounting at rate $r$.   
We want
$$
\text{PV}_{\text{refund}}
= \mathbb{E}\!\left[e^{-r\tau}\,P\!\left(1-\frac{\tau}{\Delta t}\right)\mathbf{1}_{\{\tau\le \Delta t\}}\right]
= P\int_0^{\Delta t} e^{-rt}\!\left(1-\frac{t}{\Delta t}\right) h e^{-ht}\,dt.
$$
Let $\beta = r + h$.
$$
PV_{\text{refund}}
= P \left[
h \int_0^{\Delta t} e^{-\beta t} \, dt
- \frac{h}{\Delta t} \int_0^{\Delta t} t e^{-\beta t} \, dt
\right].
$$
$$
\int_0^{\Delta t} e^{-\beta t} \, dt = \frac{1 - e^{-\beta \Delta t}}{\beta},
\qquad
\int_0^{\Delta t} t e^{-\beta t} \, dt
= \frac{1 - e^{-\beta \Delta t}(1 + \alpha \Delta t)}{\beta^2}.
$$
$$
PV_{\text{refund}}
= P \left[
\frac{h}{\beta} (1 - e^{-\beta \Delta t})
- \frac{h}{\Delta t \, \beta^2}
\left(1 - e^{-\beta \Delta t}(1 + \beta \Delta t)\right)
\right],
\qquad
\beta = r + h.
$$

<font color="#1E90FF" size="4"><b>№2 (c)</b></font>

>  What is the fair value of $P$ for this one-period default swap?

$$
P = PV_{\text{loss}} + PV_{\text{refund}}.
$$
With constant interest rate $r$ and hazard rate $h$, let $\alpha = r + h$ and $E = e^{-\beta \Delta t}$.
$$
PV_{\text{loss}} = (1 - R)\frac{h}{\beta}(1 - E),
\qquad
PV_{\text{refund}} = P \left[
\frac{h}{\beta}(1 - E)
- \frac{h}{\Delta t \, \beta^2}(1 - E(1 + \beta \Delta t))
\right].
$$
Plugging into the fair value condition and solving for $P$:
$$
P(1 - A) = PV_{\text{loss}}, \qquad
A = \frac{h}{\beta}(1 - E)
- \frac{h}{\Delta t \, \beta^2}(1 - E(1 + \beta \Delta t)).
$$
Thus,
$$
P = \frac{(1 - R)\frac{h}{\beta}(1 - E)}
{1 - \frac{h}{\beta}(1 - E)
+ \frac{h}{\Delta t \, \beta^2}(1 - E(1 + \beta \Delta t))}.
$$
Simplifying:
$$
P = (1 - R)
\frac{h \beta (1 - E)}
{\beta^2 - h\beta + \frac{h}{\Delta t}(1 - E)}.
$$
Using $\beta = r + h$ and $\beta^2 - h\beta = \beta r = r(r + h)$, we obtain the compact form:
$$
P = (1 - R)
\frac{h(r + h)\left(1 - e^{-(r + h)\Delta t}\right)}
{r(r + h) + \frac{h}{\Delta t}\left(1 - e^{-(r + h)\Delta t}\right)}.
$$

> We have a positive random variable $X$ whose distribution function is $F(x)$.

We define:
$$
S(x) = 1 - F(x)
$$
If $X$ represents a **loss distribution**, we call $S(x)$ the *distribution function for the excess loss distribution*,as it expresses the probability of losses larger than a given amount.  
If $X$ represents the **time-until-default** or **time-until-death**, we would call $S(x)$ the *survival-time distribution function*.

<font color="#1E90FF" size="4"><b>№3 (a)</b></font>

Prove that:
$$
E(X) = \int_{0}^{+\infty} S(x)\,dx
$$


Let $X \ge 0$ be a nonnegative random variable with distribution function $F(x)$  
and survival function $S(x) = 1 - F(x) = \mathbb{P}(X > x)$.  
We can write $X$ as an integral of indicator functions:
$$
X = \int_{0}^{\infty} \mathbf{1}_{\{X > t\}} \, dt
$$
Taking expectations on both sides:
$$
\mathbb{E}[X] = \mathbb{E}\!\left[\int_{0}^{\infty} \mathbf{1}_{\{X > t\}} \, dt\right]
$$
**Geometric intuition:** the area under the step function $t \mapsto \mathbf{1}_{\{X>t\}}$ from $0$ to $\infty$ is exactly the height $X$.  
Since the integrand is non-negative, we can interchange the order of expectation and integration (by Tonelli’s theorem):
$$
\mathbb{E}[X] = \int_{0}^{\infty} \mathbb{E}\!\left[\mathbf{1}_{\{X > t\}}\right] dt
$$
But $\mathbb{E}[\mathbf{1}_{\{X > t\}}] = \mathbb{P}(X > t) = S(t)$, so:
$$
\mathbb{E}[X] = \int_{0}^{\infty} S(t)\, dt
$$
$$
\mathbb{E}[X] = \int_{0}^{\infty} S(x)\, dx
$$
Alternative derivation (absolutely continuous case)   
If $X$ has a density $f$ on $[0,\infty)$, then $F'(x)=f(x)$ a.e. and $S'(x)=-f(x)$ a.e.
Start from the usual definition
$$
\mathbb{E}[X]=\int_0^\infty x\,f(x)\,dx.
$$
Integrate by parts with $u=x$ and $dv=f(x)\,dx$ (so $du=dx$ and $v=F(x)$):
$$
\int_0^\infty x\,f(x)\,dx
= \Big[x\,F(x)\Big]_{0}^{\infty} - \int_0^\infty F(x)\,dx.
$$
Since $X\ge 0$, $F(0)\ge 0$, and for integrable $X$ we have $xF(x)\to 0$ as $x\to\infty$,
so the boundary term vanishes and
$$
\mathbb{E}[X] = -\int_0^\infty F(x)\,dx.
$$
Because $S(x)=1-F(x)$,
$$
-\int_0^\infty F(x)\,dx
= \int_0^\infty (1-F(x))\,dx
= \int_0^\infty S(x)\,dx.
$$

<font color="#1E90FF" size="4"><b>№3 (b)</b></font>

> Suppose $X$ is the loss distribution, and we have an **insurance contract** with a **deductible** $L$ and **ceiling** $U$.

It pays:
- nothing if loss < $L$;
- $X - L$ if $L \le X \le U$;
- $U - L$ if $X > U$.

Prove that the expected payoff for this insurance contract is:
$$
E(D) = \int_{L}^{U} S(x)\,dx
$$

Suppose $X$ is the loss distribution, and the insurance contract pays:
$$
D =
\begin{cases}
0, & X < L,\\[4pt]
X - L, & L \le X \le U,\\[4pt]
U - L, & X > U.
\end{cases}
$$
Equivalently, $D = \min\!\big((X - L)_+,\, U - L\big)$.   
For any non-negative random variable $Y$:
$$
\mathbb{E}[Y] = \int_0^{\infty} \Pr(Y > t)\,dt.
$$
Apply this to $Y = D$:
$$
\mathbb{E}[D] = \int_0^{U - L} \Pr(D > t)\,dt,
$$
since $D \le U - L$.  
For $0 \le t < U - L$:
$$
\{D > t\} \iff \min((X - L)_+,\, U - L) > t
\iff (X - L)_+ > t
\iff X > L + t.
$$
Hence,
$$
\Pr(D > t) = \Pr(X > L + t) = S(L + t),
$$
where $S(x) = \Pr(X > x)$ is the survival function.   
Therefore,
$$
\mathbb{E}[D] = \int_0^{U - L} S(L + t)\,dt
= \int_L^U S(x)\,dx,
$$
after the change of variable $x = L + t$.   
$D = (X - L)_+ - (X - U)_+$.  
Using the identity
$$
\mathbb{E}[(X - a)_+] = \int_a^{\infty} S(x)\,dx,
$$
we obtain
$$
\mathbb{E}[D]
= \int_L^{\infty} S(x)\,dx - \int_U^{\infty} S(x)\,dx
= \int_L^U S(x)\,dx.
$$
This holds for any loss distribution $X$, without assuming continuity

> If you are a **senior tranche protection buyer**, would you prefer the underlying correlation to be **higher** or **lower**?  
How about an **equity tranche protection buyer**?

A **senior tranche protection buyer** benefits when extreme, system-wide losses occur, because the senior tranche is only affected after all junior tranches have been exhausted. When the **underlying correlation increases**, defaults tend to occur together - companies fail in clusters during economic downturns. This creates a fatter tail in the loss distribution and raises the probability that the senior tranche will suffer losses.  

**Therefore, a senior tranche protection buyer prefers higher correlation**, as it increases the chance of large, joint defaults and higher protection payouts.

In contrast, an **equity tranche protection buyer** benefits from frequent small-to-moderate losses. When correlation is **low**, defaults are more independent, leading to more dispersed, moderate loss events that frequently affect the equity tranche. When correlation is high, the outcomes are often all-or-nothing: either no defaults (no payout) or total collapse (but the protection is capped).  

**Thus, an equity tranche protection buyer prefers lower correlation**, since it produces more consistent, moderate losses and steadier protection payments.

$\varepsilon_i$’s are uncorrelated noises with mean 0 and variance 1.  
$$
\text{Cov}(X_i, X_j) = \rho \quad \text{for } i \neq j
$$
The survival time for each credit is linked to its asset return as follows:
$$
\tau_i = F_i^{-1}\!\left(N(X_i)\right)

where $F(t)$ is the survival time distribution, and $N()$ is the standard normal distribution function.  
Conditional on the common market factor $X_M$, all credit survival times are independent of each other.

**Prove that the conditional default probability**
$$

$$
can be linked to the non-conditional default probability
$$
q_i(t) = \Pr[\tau_i < t]

$$
q_i(t \mid X_M) =

where $N()$ is the standard normal distribution function, and $q_i(t) = \Pr[\tau_i < t]$.

$$

$$

Conditioning on $X_M$, the asset return is Gaussian:
$$

$$

$$

The distribution function of survival time $\tau_i$ is denoted as $F_i(t) = \Pr[\tau_i \le t]$.

$$

\;=\; \sum_{i=1}^n B_i\, \mathbb{E}\!\left[\mathbf{1}\{\tau_i < t\}\mid v\right]

Using conditional independence:
$$
\sigma_v^2 \;:=\; \mathrm{Var}\!\left(L_n(t)\mid X_M=v\right)

\;=\; \sum_{i=1}^n B_i^2\, q_i(t\mid v)\big(1 - q_i(t\mid v)\big).

> Implement Hart algorithm for one dimensional normal CDF in Python.

<font color="#1E90FF" size="4"><b>№5 (a)</b></font>

```python
import numpy as np

def hart_cdf(x):
    x = np.asarray(x, float)
    p = 0.2316419
    a1, a2, a3, a4, a5 = 0.319381530, -0.356563782, 1.781477937, -1.821255978, 1.330274429
    inv_s2pi = 1 / np.sqrt(2 * np.pi)

    ax = np.abs(x)
    t = 1 / (1 + p * ax)
    phi = inv_s2pi * np.exp(-0.5 * ax * ax)
    poly = (((a5 * t + a4) * t + a3) * t + a2) * t + a1
    cdf = 1 - phi * poly
    return np.where(x >= 0, cdf, 1 - cdf).item() if np.isscalar(x) else np.where(x >= 0, cdf, 1 - cdf)


# Hart 1D CDF
test_points = [-2.0, -1.0, 0.0, 1.0, 2.0]
print("Hart Normal CDF approximation:")
print("x\tHart(x)")
print("-" * 25)
for x in test_points:
    hart_result = hart_cdf(x)
    print(f"{x:>4.1f}\t{hart_result:.6f}")
```

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
# Standard normal PDF/CDF/PPF
def phi_pdf(x): return np.exp(-0.5*x*x)/np.sqrt(2*np.pi)

def phi_cdf(z):
    p,a1,a2,a3,a4,a5 = 0.2316419,0.319381530,-0.356563782,1.781477937,-1.821255978,1.330274429
    t=1/(1+p*np.abs(z)); poly=(((a5*t+a4)*t+a3)*t+a2)*t+a1
    c=1-phi_pdf(np.abs(z))*poly
    return np.where(z>=0,c,1-c)

def phi_ppf(p,it=60):
    p=np.asarray(p); lo,hi=-10*np.ones_like(p),10*np.ones_like(p)
    for _ in range(it):
        mid=(lo+hi)/2; sel=phi_cdf(mid)<p
        lo=np.where(sel,mid,lo); hi=np.where(sel,hi,mid)
    return 0.5*(lo+hi)

# Model Functions
def q_cond(v): return phi_cdf((phi_ppf(np.full_like(v,q))-np.sqrt(rho)*v)/np.sqrt(1-rho))

def tail_vasicek(x):
    d=np.minimum(x/(n*B),1.0)
    arg=(np.sqrt(1-rho)*phi_ppf(d)-phi_ppf(q))/np.sqrt(rho)
    return 1-phi_cdf(arg)

def tail_cn(x,m=64):
    zt,wt=np.polynomial.hermite.hermgauss(m)
    v=np.sqrt(2)*zt; qc=q_cond(v)
    Mv,sv=n*B*qc,np.sqrt(n*(B**2)*qc*(1-qc))
    tail=np.where(sv>0,1-phi_cdf((x-Mv)/sv),(Mv>x).astype(float))
    return np.sum(wt*tail)/np.sqrt(np.pi)

def binom_tail(x,p):
    pmf=np.zeros(n+1); pmf[0]=(1-p)**n; r=p/(1-p+1e-300)
    for k in range(1,n+1): pmf[k]=pmf[k-1]*((n-k+1)/k)*r
    return pmf[(np.arange(n+1)*B)>x].sum()

def tail_recursive(x,m=32):
    zt,wt=np.polynomial.hermite.hermgauss(m)
    v=np.sqrt(2)*zt; qc=q_cond(v)
    tails=np.array([binom_tail(x,p) for p in qc])
    return np.sum(wt*tails)/np.sqrt(np.pi)

# Evaluate
x_grid=np.linspace(0,n*B,40)
tails_vas=[tail_vasicek(x) for x in x_grid]
tails_cn =[tail_cn(x) for x in x_grid]
tails_rec=[tail_recursive(x) for x in x_grid]
```

The **Vasicek approximation** is the simplest and fastest. It assumes an infinitely large, homogeneous portfolio where all names have identical exposure and default probability. This gives a closed-form solution that captures the average shape of the loss distribution but slightly smooths out its discrete nature. It provides a good overall benchmark, especially in mid-loss regions, though it may deviate slightly in the tails due to finite-size effects.

The **Conditional Normal** approach improves accuracy by conditioning on the market factor and using the exact conditional mean and variance of portfolio losses. It integrates over this factor using numerical quadrature, producing a smooth and accurate estimate that closely matches the true distribution. Compared with Vasicek, CN typically yields marginally higher tail probabilities because it captures variance across realizations of the common factor.

The **Recursive (binomial mixture)** method is the most precise. It computes the exact discrete loss probabilities given the conditional default rate and then averages over the common factor. This retains the natural stepwise structure of the loss distribution and serves as the reference or truth.

In summary, under a one-factor correlation structure of 20%, all three computational methods are consistent and reliable. The Recursive method is exact but computationally heavier, Conditional Normal offers nearly identical results with much less computation, and Vasicek provides an efficient analytic limit that performs well for large, homogeneous portfolios.

>  Prove that in Bayesian statistics, the Bayesian point estimate is the posterior mean when the loss function is the square error, and the posterior median when the loss function is the absolute difference.

 In Bayesian estimation, the Bayes estimator (the point estimate that minimizes
posterior expected loss) is:  
- the **posterior mean** under **squared-error loss** $L(a,\theta) = (a - \theta)^2$  
- the **posterior median** under **absolute-error loss** $L(a,\theta) = |a - \theta|$

Let $\pi(\theta \mid x)$ be the posterior density of $\theta$ given data $x$,  
and let $a \in \mathbb{R}$ be a candidate point estimate.   
Squared-error loss $L(a,\theta) = (a - \theta)^2$

The posterior risk (posterior expected loss) is  
$$
R(a \mid x) = \mathbb{E}[(a - \theta)^2 \mid x]
= \int (a - \theta)^2 \, \pi(\theta \mid x) \, d\theta.
$$
Expand the square:
$$
R(a \mid x)
= \int (a^2 - 2a\theta + \theta^2) \, \pi(\theta \mid x) \, d\theta
= a^2 - 2a \, \mathbb{E}[\theta \mid x] + \mathbb{E}[\theta^2 \mid x].
$$
As a function of $a$, this is a strictly convex quadratic. Differentiate and set to zero:
$$
\frac{d}{da} R(a \mid x) = 2a - 2\mathbb{E}[\theta \mid x] = 0
\quad \Rightarrow \quad a^\star = \mathbb{E}[\theta \mid x].
$$
Since $\frac{d^2 R}{da^2} = 2 > 0$, this minimizer is unique.  
The Bayes estimator is the **posterior mean**.  
Absolute-error loss $L(a,\theta) = |a - \theta|$  
The posterior risk is  
$$
R(a \mid x) = \mathbb{E}[|a - \theta| \mid x]
= \int |a - \theta| \, \pi(\theta \mid x) \, d\theta.
$$
Split the integral at $a$:
$$
R(a \mid x)
= \int_{-\infty}^{a} (a - \theta) \, \pi(\theta \mid x) \, d\theta
+ \int_{a}^{\infty} (\theta - a) \, \pi(\theta \mid x) \, d\theta.
$$
Differentiate with respect to $a$ (using Leibniz’s rule and that $|a - \theta|$ is almost everywhere differentiable):
$$
\frac{d}{da} R(a \mid x)
= \int_{-\infty}^{a} \pi(\theta \mid x) \, d\theta
- \int_{a}^{\infty} \pi(\theta \mid x) \, d\theta
= 2F(a \mid x) - 1,
$$
where $F(a \mid x) = \int_{-\infty}^{a} \pi(\theta \mid x) \, d\theta$ is the posterior cumulative distribution function (CDF).
The optimal $a^\star$ satisfies
$$
2F(a^\star \mid x) - 1 = 0
\quad \Rightarrow \quad F(a^\star \mid x) = \tfrac{1}{2}.
$$

     {\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}},
\qquad
\bar{x} = \frac{1}{n}\sum_{i=1}^n x_i.
$$
Because $x_{n+1} \mid \mu \sim \mathcal{N}(\mu, \sigma^2)$,  
the posterior predictive mean equals the posterior mean of $\mu$:
$$
\
\mathbb{E}[x_{n+1} \mid x_{1:n}]
= \mu_n
= w\,\bar{x} + (1 - w)\,\mu_0
\
$$
where the **data weight** is
$$
\
w =
\frac{\frac{n}{\sigma^2}}
     {\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}}
= \frac{n\,\sigma_0^2}{n\,\sigma_0^2 + \sigma^2}.
\
$$

<font color="#1E90FF" size="4"><b>№8 (b)</b></font>

> Provide an intuitive explanation for the weight, and explain each item in the weight, and how it impacts the final result.


We have:
$$
E[x_{n+1} \mid x_1, x_2, \dots, x_n] = \mu_n = w \, \bar{x} + (1 - w)\, \mu_0,
$$
where
$$
w = \frac{n \, \sigma_0^2}{n \, \sigma_0^2 + \sigma^2}.
$$
Meaning of Each Symbol
| Symbol | Meaning | Role |
|:--------|:---------|:------|
| $n$ | Sample size | More data → stronger influence of sample mean $ \bar{x} $ |
| $\sigma^2$ | Observation variance | Higher noise → weaker confidence in data |
| $\sigma_0^2$ | Prior variance of $ \mu $ | Larger prior uncertainty → weaker prior influence |
| $\mu_0$ | Prior mean | Prior belief about the unknown mean before observing data |
| $\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i$ | Sample average | Information extracted from the data |
| $\mu_n$ | Posterior mean | Weighted combination of prior and data |
| $w$ | Data weight | Relative confidence in the sample mean |

The weight $w$ determines **how much we trust the data versus the prior**:
$$
\
w =
\frac{\frac{n}{\sigma^2}}
     {\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}}
= \frac{n\,\sigma_0^2}{n\,\sigma_0^2 + \sigma^2}.
\
$$
- **If $w \approx 1$:**  
Many or precise observations ($n \uparrow$, $\sigma^2 \downarrow$)  
Posterior mean $ \mu_n \approx \bar{x} $ → data dominate.

- **If $w \approx 0$:**  
Few or noisy observations ($n \downarrow$, $\sigma^2 \uparrow$)  
Posterior mean $ \mu_n \approx \mu_0 $ → prior dominates.

How Each Term Affects $w$
| Factor | Effect on $w$ | Interpretation |
|:--------|:------------------|:----------------|
| $n \uparrow$ | $w \uparrow$ | More data. Therefore, rely more on observations |
| $\sigma^2 \uparrow$ | $w \downarrow$ | More noise. Therefore, rely more on prior |
| $\sigma_0^2 \uparrow$ | $w \uparrow$ | Weaker prior. Therefore, data gain influence |

<font color="#1E90FF" size="4"><b>№9</b></font>

> Suppose $g(x, \theta)$ is an **unbiased estimating function** (whose expectation is zero with respect to the true distribution) for parameter $\theta$ in the model $f(x, \theta)$, which is the density function.

<font color="#1E90FF" size="4"><b>№9 (a)</b></font>

$E\!\left[\dfrac{\partial \log f}{\partial \theta}\right] = 0$

We show that the expected value of the score function is zero.
$$
E_\theta\!\left[\frac{\partial}{\partial\theta}\log f(X,\theta)\right]
= \int \frac{\partial}{\partial\theta}\log f(x,\theta)\, f(x,\theta)\,dx
$$
$$
= \int \frac{1}{f(x,\theta)}\,\frac{\partial f(x,\theta)}{\partial\theta}\, f(x,\theta)\,dx
$$
$$
= \int \frac{\partial f(x,\theta)}{\partial\theta}\,dx
$$
$$
= \frac{\partial}{\partial\theta}\int f(x,\theta)\,dx
$$
$$
= \frac{\partial}{\partial\theta}(1)
$$
$$
= 0
$$

<font color="#1E90FF" size="4"><b>№9 (b)</b></font>

$E\!\left[\dfrac{\partial^2 \log f}{\partial \theta^2}\right] = -E\!\left[\left(\dfrac{\partial \log f}{\partial \theta}\right)^2\right]$

Let the **score** be $U(X,\theta)=\dfrac{\partial}{\partial \theta}\log f(X,\theta)$.  
Then $E[U]=0$ (from part a).

Differentiate $E[U]=0$ w.r.t. $\theta$:
$$
0=\frac{d}{d\theta}E[U]
=E\!\left[\frac{\partial U}{\partial \theta}\right]+E\!\left[U\,\frac{\partial \log f}{\partial \theta}\right].
$$

But $\dfrac{\partial f}{\partial \theta}=f\,U$, so the second term is $E[U^2]$. Hence,
$$
E\!\left[\frac{\partial^2 \log f}{\partial \theta^2}\right]
= -E[U^2]
= -E\!\left[\left(\frac{\partial \log f}{\partial \theta}\right)^2\right].
$$

Thus,
$$
\mathbb{E}_\theta[U(X,\theta)] = 0.
$$
Differentiate both sides of
$$
\mathbb{E}_\theta[U(X,\theta)] = 0
$$
with respect to $\theta$. The left-hand side becomes:
$$
\frac{d}{d\theta} \mathbb{E}_\theta[U(X,\theta)]
= \frac{d}{d\theta} \int U(x,\theta) f(x;\theta)\, dx.
$$
We apply product rule inside the integral:
$$
\frac{d}{d\theta} \int U f \, dx
= \int \frac{\partial U}{\partial \theta} f \, dx
+ \int U \frac{\partial f}{\partial \theta} \, dx.
$$
So
$$
0
= \frac{d}{d\theta} \mathbb{E}_\theta[U]
= \int \frac{\partial U}{\partial \theta} f \, dx
+ \int U \frac{\partial f}{\partial \theta} \, dx.
$$
Rewrite each term as an expectation. The first term:
$$
\int \frac{\partial U}{\partial \theta} f \, dx
= \mathbb{E}_\theta\!\left[\frac{\partial U}{\partial \theta}\right].
$$
For the second term, observe that
$$
\frac{\partial f}{\partial \theta}
= f \cdot \frac{\partial \log f}{\partial \theta}
= f \cdot U.
$$
Then
$$
\int U \frac{\partial f}{\partial \theta} \, dx
= \int U \big( f \cdot U \big) \, dx
= \int U^2 f \, dx
= \mathbb{E}_\theta[U^2].
$$
Therefore,
$$
0
= \mathbb{E}_\theta\!\left[\frac{\partial U}{\partial \theta}\right]
+ \mathbb{E}_\theta[U^2].
$$
Equivalently,
$$
\mathbb{E}_\theta\!\left[\frac{\partial U}{\partial \theta}\right]
= -\,\mathbb{E}_\theta[U^2].
$$
Express in log-likelihood notation
$$
U(X,\theta)
= \frac{\partial}{\partial \theta} \log f(X;\theta).
$$
Then
$$
\frac{\partial U}{\partial \theta}
= \frac{\partial^2}{\partial \theta^2} \log f(X;\theta).
$$
So the identity becomes
$$
\mathbb{E}_\theta \!\left[ \frac{\partial^2}{\partial \theta^2} \log f(X;\theta) \right]
= - \mathbb{E}_\theta \!\left[
\left( \frac{\partial}{\partial \theta} \log f(X;\theta) \right)^2
\right].
$$
The left-hand side is the expected second derivative (the expected curvature) of the log-likelihood.  
The right-hand side is the Fisher information:
$$
\mathcal{I}(\theta)
= \mathbb{E}_\theta \!\left[
\left( \frac{\partial}{\partial \theta} \log f(X;\theta) \right)^2
\right].
$$
So we can also write
$$
\mathbb{E}_\theta \!\left[ \frac{\partial^2}{\partial \theta^2} \log f(X;\theta) \right]
= - \mathcal{I}(\theta).
$$

<font color="#1E90FF" size="4"><b>№9 (c)</b></font>

$\displaystyle \frac{E[g^2]}{(E[\partial g / \partial \theta])^2} \ge \frac{1}{E\!\left[\left(\dfrac{\partial \log f}{\partial \theta}\right)^2\right]}$

Since $g(x,\theta)$ is unbiased, $E[g]=0$. Differentiate this expectation:

$$
0 = \frac{d}{d\theta}E[g]
= \int \left(\frac{\partial g}{\partial \theta}f + g\,\frac{\partial f}{\partial \theta}\right)dx
= E\!\left[\frac{\partial g}{\partial \theta}\right] + E[gU],
$$

where $U = \dfrac{\partial \log f}{\partial \theta}$. Thus,

$$
E[gU] = -E\!\left[\frac{\partial g}{\partial \theta}\right].
$$

Apply the Cauchy–Schwarz inequality:

$$
(E[gU])^2 \le E[g^2]\,E[U^2]
\quad\Rightarrow\quad
\big(E[\tfrac{\partial g}{\partial \theta}]\big)^2
\le E[g^2]\,E[U^2].
$$

Rearrange:

$$
E[g^2] \ge \frac{\big(E[\tfrac{\partial g}{\partial \theta}]\big)^2}{E[U^2]}
= \frac{\big(E[\tfrac{\partial g}{\partial \theta}]\big)^2}{E\!\left[\left(\tfrac{\partial \log f}{\partial \theta}\right)^2\right]}.
$$

Equivalently,

$$
\frac{E[g^2]}{\big(E[\tfrac{\partial g}{\partial \theta}]\big)^2}
\ge \frac{1}{E\!\left[\left(\tfrac{\partial \log f}{\partial \theta}\right)^2\right]}.
$$

The denominator on the right is the **Fisher information**

$$
I(\theta) = E\!\left[\left(\dfrac{\partial \log f}{\partial \theta}\right)^2\right].
$$

Thus, this is the **information inequality** for unbiased estimating functions.
