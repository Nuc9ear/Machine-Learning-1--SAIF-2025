In a general supervised machine learning problem we try to find a function $g(x)$ to predict $Y$ based on a set of variables $X$. One of the predictors is the conditional expectation:
$$
f^*(x) = E(Y \mid X)
$$

Prove that it is the best predictor based on the squared loss function. That is, for any $g(x)$, we have

$$
E[(Y - f^*(X))^2] \le E[(Y - g(X))^2]
$$

We want to show that for any measurable function $g(X)$,
$$
\mathbb{E}\big[(Y - f^*(X))^2\big] \le \mathbb{E}\big[(Y - g(X))^2\big],
$$
where
$$
f^*(X) = \mathbb{E}[Y|X].
$$
We can rewrite:
$$
Y - g(X) = (Y - f^*(X)) + (f^*(X) - g(X)).
$$
Then, by expanding the square:
$$
(Y - g(X))^2 = (Y - f^*(X))^2
+ 2(Y - f^*(X))(f^*(X) - g(X))
+ (f^*(X) - g(X))^2.
$$
Taking expectations and using linearity:
$$
\mathbb{E}\!\left[(Y - g(X))^2\right]
= \mathbb{E}\!\left[(Y - f^*(X))^2\right]
+ 2\,\mathbb{E}\!\left[(Y - f^*(X))(f^*(X) - g(X))\right]
+ \mathbb{E}\!\left[(f^*(X) - g(X))^2\right].
$$
We have cross-term:
$$
\mathbb{E}\!\left[(Y - f^*(X))(f^*(X) - g(X))\right]
= \mathbb{E}\!\left[\mathbb{E}\!\left[(Y - f^*(X))(f^*(X) - g(X)) \mid X\right]\right].
$$
Since $(f^*(X) - g(X))$ depends only on $X$, it can be taken out of the inner conditional expectation:
$$
= \mathbb{E}\!\left[(f^*(X) - g(X)) \, \mathbb{E}[Y - f^*(X) \mid X]\right].
$$
But by the definition of conditional expectation,
$$
\mathbb{E}[Y - f^*(X) \mid X] = \mathbb{E}[Y \mid X] - f^*(X) = 0.
$$
Hence,
$$
\begin{aligned}
\mathbb{E}\big[(Y - f^*(X))(f^*(X) - g(X))\big]
&= \mathbb{E}\Big[\mathbb{E}\big[(Y - f^*(X))(f^*(X) - g(X)) \mid X\big]\Big] \\
&= \mathbb{E}\Big[(f^*(X) - g(X))\,\mathbb{E}[Y - f^*(X) \mid X]\Big] \\
&= \mathbb{E}\big[(f^*(X) - g(X)) \cdot 0\big] = 0.
\end{aligned}
$$
Therefore,
$$
\mathbb{E}\big[(Y - g(X))^2\big]
= \mathbb{E}\big[(Y - f^*(X))^2\big]
+ \mathbb{E}\big[(f^*(X) - g(X))^2\big].
$$

Since $\mathbb{E}\big[(f^*(X) - g(X))^2\big] \ge 0$, we get

$$
\mathbb{E}\big[(Y - g(X))^2\big]
\ge
\mathbb{E}\big[(Y - f^*(X))^2\big].
$$

Equality holds **if and only if**
$$
\mathbb{E}\big[(f^*(X) - g(X))^2\big] = 0,
$$

Suppose we have the standard Wiener process $W(t)$, the following is called an integrated Wiener process

$$
\int_0^t W(s)\, ds.
$$

What is the variance of this integrated process?

Let $W(t)$ be a standard Wiener process (Brownian motion) with  
$\mathbb{E}[W(t)] = 0, \quad \text{and} \quad \operatorname{Cov}(W(s), W(u)) = \min(s,u).$
Define the **integrated Wiener process**:
$$I_t = \int_0^t W(s)\,ds.$$
Since $\mathbb{E}[I_t] = 0$,
$$
\operatorname{Var}(I_t)
= \mathbb{E}[I_t^2]
= \mathbb{E}\!\left[\int_0^t \!\int_0^t W(s)W(u)\,ds\,du \right].
$$
By linearity of expectation,
$$
\operatorname{Var}(I_t)
= \int_0^t \!\int_0^t \mathbb{E}[W(s)W(u)]\,ds\,du
= \int_0^t \!\int_0^t \min(s,u)\,ds\,du.
$$
Split the square domain into two regions ($s<u$ and $s>u$):
$$
\int_0^t\!\int_0^t \min(s,u)\,ds\,du
= \int_0^t \!\left( \int_0^u s\,ds + \int_u^t u\,ds \right)\,du.
$$
Compute each part:
$$
\int_0^u s\,ds = \frac{u^2}{2}, \qquad
\int_u^t u\,ds = u(t-u).
$$
Substitute:
$$
\int_0^t \!\left( \frac{u^2}{2} + u(t-u) \right)\,du
= \int_0^t \!\left( ut - \frac{u^2}{2} \right)\,du.
$$
Integrate:
$$
\left[ \frac{u^2 t}{2} - \frac{u^3}{6} \right]_0^t
= \frac{t^3}{2} - \frac{t^3}{6}
= \frac{t^3}{3}.
$$
Hence,
$$
{\operatorname{Var}\!\left( \int_0^t W(s)\,ds \right) = \frac{t^3}{3}.}
$$
Alternative Proof via **Itô Isometry**  
Using integration by parts:
$$
\int_0^t W(s)\,ds = \int_0^t (t-s)\,dW(s).
$$
Then by Itô isometry:
$$
\operatorname{Var}(I_t)
= \mathbb{E}\!\left[\left(\int_0^t (t-s)\,dW(s)\right)^2\right]
= \int_0^t (t-s)^2\,ds
= \frac{t^3}{3}.
$$

We look at a process $X(t)$ over a time interval $[0, t]$ and study how to approximate this stochastic process using a random walk model. We divide the time interval into many small time steps $\Delta t = \frac{t}{n}$. At each time unit we have a symmetric random walk, moving up or down by $\Delta x$ unit with equal probability $1/2$. That is, $Y_i$ has a value $1$ if it moves up and a value of $-1$ if it moves down.

After $n$ steps, the position of process $X_t$ could be written as follows
$$
X(t) = \Delta x (Y_1 + Y_2 + \dots + Y_{\left[ \frac{t}{\Delta t} \right]})
$$
where $\left[ \frac{t}{\Delta t} \right]$ is the largest integer less than or equal to $\frac{t}{\Delta t}$, and the $Y_i$’s are assumed independent with
$$
Pr[Y_i = 1] = Pr[Y_i = −1] = \frac{1}{2}.
$$

<font color="#1E90FF" size="4"><b>№3 (a)</b></font>

Prove
$$
E(X(t)) = 0
$$
$$
Var(X(t)) = (\Delta x)^2 \left\{ \frac{t}{\Delta t} \right\}
$$

We consider a symmetric random walk approximation to a stochastic process $X(t)$ on the interval $[0, t]$.  
We divide the interval into small steps of size $\Delta t$ and define $m = \left\lfloor \frac{t}{\Delta t} \right\rfloor$.    
At each time step, the process moves **up or down** by $\Delta x$ with equal probability $1/2$:
$$
Y_i =
\begin{cases}
+1, & \text{with probability } \tfrac{1}{2}, \\
-1, & \text{with probability } \tfrac{1}{2}.
\end{cases}
$$
The position after $m$ steps is:
$$
X(t) = \Delta x (Y_1 + Y_2 + \cdots + Y_m)
= \Delta x \sum_{i=1}^{m} Y_i.
$$
Each $Y_i$ is independent and identically distributed with:
$$
\mathbb{E}[Y_i] = 0, \qquad \operatorname{Var}(Y_i) = 1.
$$
Using linearity of expectation:
$$
\mathbb{E}[X(t)] = \Delta x \sum_{i=1}^{m} \mathbb{E}[Y_i]
= \Delta x \cdot m \cdot 0
= 0.
$$
Thus:
$$
{\mathbb{E}[X(t)] = 0.}
$$
Since the $Y_i$’s are independent:
$$
\operatorname{Var}(X(t))
= \operatorname{Var}\!\left( \Delta x \sum_{i=1}^{m} Y_i \right)
= (\Delta x)^2 \operatorname{Var}\!\left(\sum_{i=1}^{m} Y_i\right)
= (\Delta x)^2 \sum_{i=1}^{m} \operatorname{Var}(Y_i).
$$
Each $\operatorname{Var}(Y_i) = 1$, so:
$$
\operatorname{Var}(X(t)) = (\Delta x)^2 \, m.
$$
Substituting $m = \left\lfloor \frac{t}{\Delta t} \right\rfloor$:
$$
\operatorname{Var}(X(t)) = (\Delta x)^2 \left\lfloor \frac{t}{\Delta t} \right\rfloor.
$$

<font color="#1E90FF" size="4"><b>№3 (b)</b></font>

Under which condition that we would have the following? How do you state this conditional in a simple non-math language?
$$
E(X(t)) = 0
$$
$$
Var(X(t)) \rightarrow \sigma^2 t
$$

We start with the discrete random walk model:
$$
X(t) = \Delta x \sum_{i=1}^{m} Y_i,
\qquad m = \frac{t}{\Delta t},
$$
where each $Y_i$ is independent with  
$\mathbb{E}[Y_i] = 0$ and $\operatorname{Var}(Y_i) = 1$.    
We already have:
$$
\mathbb{E}[X(t)] = 0, \qquad
\operatorname{Var}(X(t)) = (\Delta x)^2 \frac{t}{\Delta t}.
$$
For the random walk to approximate a continuous process (Brownian motion) with variance rate $\sigma^2$, we require that as the time steps get smaller:
$$
\operatorname{Var}(X(t)) \to \sigma^2 t.
$$
This means that:
$$
\lim_{\Delta t \to 0} \frac{(\Delta x)^2}{\Delta t} = \sigma^2.
$$
or equivalently,
$$
\Delta x = \sigma \sqrt{\Delta t}.
$$
Intuitive (non-mathematical) explanation:
> As we make the time steps smaller, we must also make the jump size smaller so that **the average squared movement per unit time remains constant**.  

> The step size $\Delta x$ should shrink proportionally to $\sqrt{\Delta t}$.

This ensures that:
- The process doesn’t “explode” (variance stays finite),
- The random walk becomes **continuous** in the limit,
- And the limiting process is a **Brownian motion** with volatility $\sigma$.

Under the scaling condition $\Delta x = \sigma \sqrt{\Delta t}$,
$$
\mathbb{E}[X(t)] = 0, \qquad
\operatorname{Var}(X(t)) \to \sigma^2 t,
$$
so the random walk converges in distribution to a **Brownian motion** with variance parameter $\sigma^2$.

<font color="#1E90FF" size="4"><b>№4</b></font>

We have observed $n$ observations $x_1, x_2, \ldots, x_n$ at times $t_1, t_2, \ldots, t_n$ from a Brownian motion with a constant drift $\mu$ and a constant variation $\sigma$. Find the maximum likelihood estimators for $\mu$ and $\sigma$.

We observe $x_1, x_2, \dots, x_n$ at times $t_1, t_2, \dots, t_n$ from a Brownian motion with drift $\mu$ and variance parameter $\sigma^2$. That is:
$$
X(t) \sim \text{Brownian motion with drift } \mu, \text{ diffusion coefficient } \sigma.
$$
Meaning:
$$
dX(t) = \mu\,dt + \sigma\,dW(t),
$$
where $W(t)$ is a standard Wiener process.  
Equivalently:
$$
X(t_i) - X(t_{i-1}) \sim \mathcal{N}\big(\mu (t_i - t_{i-1}),\, \sigma^2 (t_i - t_{i-1})\big),
$$
and increments are independent.   
Let $\Delta t_i = t_i - t_{i-1}$ for $i = 1, \dots, n$, with $t_0 = 0$ and possibly $X(0) = x_0$ known.   
Then:
$$
X(t_1) \sim \mathcal{N}(x_0 + \mu t_1,\, \sigma^2 t_1).
$$
It is better to work with **increments**:
$$
y_i = x_i - x_{i-1}, \qquad i = 1, \dots, n.
$$
If $x_0$ is given, or even if not, we can estimate $\mu$ and $\sigma$ from the increments.    
The joint density of $(x_1, \dots, x_n)$ given $x_0$ is:
$$
f(x_1, \dots, x_n \mid x_0, \mu, \sigma)
= \prod_{i=1}^{n}
\frac{1}{\sqrt{2\pi \sigma^2 \Delta t_i}}
\exp\!\left[-\frac{(x_i - x_{i-1} - \mu \Delta t_i)^2}{2\sigma^2 \Delta t_i}\right].
$$
Let $\Delta x_i = x_i - x_{i-1}$ and $\Delta t_i = t_i - t_{i-1}$.  
Then the log-likelihood is:
$$
\ell(\mu, \sigma^2)
= -\frac{n}{2}\log(2\pi)
- \frac{n}{2}\log(\sigma^2)
- \frac{1}{2}\sum_{i=1}^{n}\log(\Delta t_i)
- \frac{1}{2\sigma^2}\sum_{i=1}^{n}
\frac{(\Delta x_i - \mu \Delta t_i)^2}{\Delta t_i}.
$$
Differentiate with respect to $\mu$:
$$
\frac{\partial \ell}{\partial \mu}
= \frac{1}{\sigma^2}\sum_{i=1}^{n}(\Delta x_i - \mu \Delta t_i) = 0.
$$
Solve for $\mu$:
$$
\sum_{i=1}^{n} \Delta x_i - \mu \sum_{i=1}^{n} \Delta t_i = 0,
$$
hence:
$$
\hat{\mu} = \frac{\sum_{i=1}^{n} \Delta x_i}{\sum_{i=1}^{n} \Delta t_i}.
$$
But note that:
$$
\sum_{i=1}^{n} \Delta x_i = x_n - x_0, \quad
\sum_{i=1}^{n} \Delta t_i = t_n - t_0.
$$
Therefore:
$$
\hat{\mu} = \frac{x_n - x_0}{t_n - t_0}.
$$
Differentiate $\ell$ with respect to $\sigma^2$:    
Let $\theta = \sigma^2$. Then:
$$
\frac{\partial \ell}{\partial \theta}
= -\frac{n}{2\theta}
+ \frac{1}{2\theta^2}\sum_{i=1}^{n}\frac{(\Delta x_i - \mu \Delta t_i)^2}{\Delta t_i}.
$$
Set to zero:
$$
-\frac{n}{2\theta} + \frac{1}{2\theta^2}
\sum_{i=1}^{n}\frac{(\Delta x_i - \mu \Delta t_i)^2}{\Delta t_i} = 0.
$$
Simplify:
$$
n\theta = \sum_{i=1}^{n}\frac{(\Delta x_i - \mu \Delta t_i)^2}{\Delta t_i}.
$$
Hence the MLE is:
$$
\hat{\sigma}^2
= \frac{1}{n}\sum_{i=1}^{n}\frac{(\Delta x_i - \hat{\mu}\,\Delta t_i)^2}{\Delta t_i}.
$$
$$
\hat{\mu} = \frac{x_n - x_0}{t_n - t_0}, \qquad
\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^{n}
\frac{(\Delta x_i - \hat{\mu}\,\Delta t_i)^2}{\Delta t_i}.
$$
These are the **maximum likelihood estimators** for the drift $\mu$ and the diffusion coefficient $\sigma^2$ of a Brownian motion with constant drift and variance, given discrete observations at possibly irregular times.

For a discrete probability density function $f(x)$ for a random variable which takes values on a set $(x_1, x_2, \ldots, x_n)$, the entropy is defined as
$$
H(X) = E[-\ln(f(x))] = -\sum_{i=1}^{n} f(x_i)\ln[f(x_i)]
$$
For a continuous random variable $X$ with density function $f(x)$, the entropy is defined as
$$
H(X) = E[-\ln(f(x))] = -\int f(x)\ln[f(x)]\,dx
$$

<font color="#1E90FF" size="4"><b>№6 (a)</b></font>

Using the discrete entropy definition with $n$ possible outcomes to show that the uniform distribution has the maximum entropy.

Let  
$ p_i = \Pr[X = x_i], \quad p_i \ge 0, \quad \sum_{i=1}^n p_i = 1. $    
Entropy (using natural logs) is defined as
$$
H(p) = -\sum_{i=1}^n p_i \ln p_i
$$
Lagrangian
$$
L(p, \lambda) = -\sum_{i=1}^n p_i \ln p_i + \lambda \left(\sum_{i=1}^n p_i - 1\right)
$$
First-Order Conditions (for any $p_j > 0$)
$$
\frac{\partial L}{\partial p_j} = -(\ln p_j + 1) + \lambda = 0
$$
Hence  
$$
\ln p_j = \lambda - 1 \quad \Rightarrow \quad p_j = c \quad (\text{same for all } j)
$$
Using the constraint $\sum_i p_i = 1$ gives  
$$
c = \frac{1}{n}
$$
Thus, the only stationary point in the simplex’s interior is  
$$
p_1 = p_2 = \cdots = p_n = \frac{1}{n}
$$
The entropy over $n$ outcomes is **maximized uniquely** by the **uniform distribution**  
$$
p_i = \frac{1}{n}, \quad \forall i.
$$

Suppose that a discrete random variable has three possible outcomes, and whose mean is $2.5$, what is the maximum entropy distribution with this mean constraint?

We want to maximize the entropy  
$$
H(p) = -\sum_{i=1}^3 p_i \ln p_i
$$
subject to the constraints  
$$
p_1 + p_2 + p_3 = 1, \qquad
\mathbb{E}[X] = 1\cdot p_1 + 2\cdot p_2 + 3\cdot p_3 = 2.5, \qquad
p_i \ge 0.
$$
Introduce multipliers $\mu,\lambda$ and form  
$$
L = -\sum_{i=1}^3 p_i \ln p_i
  + \mu\!\left(\sum_{i=1}^3 p_i - 1\right)
  + \lambda\!\left(\sum_{i=1}^3 i\,p_i - 2.5\right).
$$
First-Order Conditions (for each $i$ with $p_i>0$)
$$
\frac{\partial L}{\partial p_i} = -(\ln p_i + 1) + \mu + \lambda i = 0
\quad \Longrightarrow \quad
\ln p_i = \mu - 1 + \lambda i
\quad \Longrightarrow \quad
p_i \propto e^{\lambda i}.
$$
Thus the maximizing distribution has exponential form  
$$
p_i = \frac{e^{\lambda i}}{Z},
\qquad
Z = e^{\lambda} + e^{2\lambda} + e^{3\lambda}.
$$
The given mean $2.5$ implies  
$$
\frac{e^{\lambda} + 2e^{2\lambda} + 3e^{3\lambda}}{e^{\lambda} + e^{2\lambda} + e^{3\lambda}} = 2.5.
$$
Let $u = e^{\lambda} > 0$. Then  
$$
\frac{u + 2u^2 + 3u^3}{u + u^2 + u^3} = 2.5
\quad \Longleftrightarrow \quad
u^2 - u - 3 = 0
\quad \Longrightarrow \quad
u = \frac{1+\sqrt{13}}{2}.
$$
Divide numerator and denominator by $u$:
$$
p_1 = \frac{1}{1+u+u^2}, \qquad
p_2 = \frac{u}{1+u+u^2}, \qquad
p_3 = \frac{u^2}{1+u+u^2}.
$$
Using $u = \tfrac{1+\sqrt{13}}{2}$, we find  
$$
1+u+u^2 = 5 + \sqrt{13}.
$$
Final Closed-Form Probabilities
$$
p_1 = \frac{1}{5+\sqrt{13}}, \qquad
p_2 = \frac{\tfrac{1+\sqrt{13}}{2}}{5+\sqrt{13}}, \qquad
p_3 = \frac{\big(\tfrac{1+\sqrt{13}}{2}\big)^2}{5+\sqrt{13}}.

Using $\sqrt{13}\approx 3.605551$:

The ridge regression model could be taken as the ordinary regression model
$$
Y = X\beta + \varepsilon
$$
with a constraint on the regression coefficient such that the sum of the squared coefficients is minimized. That is, the ridge loss function is to minimize the following:
$$
L(\beta) = \|Y - X\beta\|^2 + \lambda \|\beta\|^2
$$
where $Y$ is an $n \times 1$ matrix, $X$ is an $n \times p$ matrix, and $\varepsilon$ is $n \times 1$ representing the errors, which are independent and have a constant variance $\sigma^2$ for all errors, $\varepsilon_1,  \varepsilon_2, \ldots, \varepsilon_n$.

<font color="#1E90FF" size="4"><b>№7 (a)</b></font>

Prove that the ridge estimator  
$$
\hat{\beta}_{ridge} = (X^T X + \lambda I_{pp})^{-1} X^T Y
$$
is as shown above, and show that it is the solution to minimize the ridge loss function.

Let
$$
L(\beta)=\lVert Y-X\beta\rVert^2+\lambda\lVert\beta\rVert^2
= (Y-X\beta)^\top (Y-X\beta)+\lambda\,\beta^\top\beta.
$$
Expand the quadratic:
$$
\begin{aligned}
L(\beta)
&= Y^\top Y - 2\beta^\top X^\top Y + \beta^\top X^\top X \beta + \lambda\,\beta^\top\beta \\
&= Y^\top Y - 2\beta^\top X^\top Y + \beta^\top \bigl(X^\top X+\lambda I_{pp}\bigr)\beta.
\end{aligned}
$$
Take the gradient and set to zero (first-order condition):
$$
\nabla_\beta L(\beta)
= -2X^\top Y + 2\bigl(X^\top X+\lambda I_{pp}\bigr)\beta
= 0
\;\;\Longrightarrow\;\;
\bigl(X^\top X+\lambda I_{pp}\bigr)\beta = X^\top Y.
$$
For $\lambda>0$, $X^\top X+\lambda I_{pp}$ is symmetric positive definite and hence invertible.   
Therefore,
$$
\hat\beta_{\text{ridge}}
=\bigl(X^\top X+\lambda I_{pp}\bigr)^{-1}X^\top Y.
$$

<font color="#1E90FF" size="4"><b>№7 (b)</b></font>

Obtain the expected value of the ridge estimator, and show that it is not an unbiased estimator.

Model:
$$
Y=X\beta+\varepsilon,\qquad \mathbb E[\varepsilon\mid X]=0,\qquad \mathrm{Var}(\varepsilon\mid X)=\sigma^2 I_n.
$$
Ridge estimator:
$$
\hat\beta_{\text{ridge}}=(X^\top X+\lambda I_p)^{-1}X^\top Y,\qquad \lambda>0.
$$
Plug the model into the estimator:
$$
\begin{aligned}
\hat\beta_{\text{ridge}}
&=(X^\top X+\lambda I)^{-1}X^\top(X\beta+\varepsilon)\\
&=(X^\top X+\lambda I)^{-1}X^\top X\,\beta \;+\; (X^\top X+\lambda I)^{-1}X^\top \varepsilon.
\end{aligned}
$$
Take the conditional expectation    
Using $\mathbb E[\varepsilon\mid X]=0$,
$$
\mathbb E[\hat\beta_{\text{ridge}}\mid X]
=(X^\top X+\lambda I)^{-1}X^\top X\,\beta
=: S_\lambda\,\beta,
$$
where
$$
S_\lambda \equiv (X^\top X+\lambda I)^{-1}X^\top X.
$$
we can rewrite $S_\lambda$ as
$$
S_\lambda = I - \lambda(X^\top X+\lambda I)^{-1}.
$$

Hence the **bias** of the ridge estimator conditional on $X$ is
$$
\begin{aligned}
\mathrm{Bias}(\hat\beta_{\text{ridge}}\mid X)
&=\mathbb E[\hat\beta_{\text{ridge}}\mid X]-\beta\\
&=(S_\lambda - I)\beta\\
&=-\lambda(X^\top X+\lambda I)^{-1}\beta.
\end{aligned}
$$

For $\lambda>0$, this bias equals zero only in trivial cases such as $\beta=0$.  
Therefore, **ridge regression is not an unbiased estimator** of $\beta$.


<font color="#1E90FF" size="4"><b>№7 (c)</b></font>

Obtain the variance of the ridge estimator.

Model:
$$
Y = X\beta + \varepsilon, \qquad \mathbb{E}[\varepsilon \mid X] = 0, \qquad \mathrm{Var}(\varepsilon \mid X) = \sigma^2 I_n.
$$
Ridge estimator:
$$
\hat{\beta}_{\text{ridge}} = (X^\top X + \lambda I)^{-1} X^\top Y.
$$

**Covariance (given $X$):**
$$
\begin{aligned}
\mathrm{Var}(\hat{\beta}_{\text{ridge}} \mid X)
&= \mathrm{Var}\!\left((X^\top X + \lambda I)^{-1} X^\top Y \; \middle| \; X \right) \\[4pt]
&= (X^\top X + \lambda I)^{-1} X^\top \, \mathrm{Var}(Y \mid X) \, X \, (X^\top X + \lambda I)^{-1} \\[4pt]
&= \sigma^2 (X^\top X + \lambda I)^{-1} X^\top X (X^\top X + \lambda I)^{-1}.
\end{aligned}
$$
Since $X^\top X + \lambda I$ is symmetric positive definite for $\lambda > 0$, its inverse is also symmetric.

<font color="#1E90FF" size="4"><b>№7 (d)</b></font>

Show that the variance of the ridge estimator is less than the variance of the maximum likelihood estimator for the ordinary regression model.

**OLS / MLE:**
$$
\mathrm{Var}(\hat{\beta}_{\text{OLS}} \mid X) = \sigma^2 A^{-1},
$$
where $A = X^\top X$.     
**Ridge (with $\lambda > 0$):**
$$
\mathrm{Var}(\hat{\beta}_{\text{ridge}} \mid X)
= \sigma^2 (A + \lambda I)^{-1} A (A + \lambda I)^{-1}.
$$
Let $A = Q D Q^\top$ with $Q$ orthogonal and  
$D = \mathrm{diag}(d_1, \dots, d_p)$, where $d_i > 0$.    
Then
$$
\mathrm{Var}(\hat{\beta}_{\text{OLS}} \mid X)
= \sigma^2 Q \, \mathrm{diag}\!\left(\frac{1}{d_i}\right) Q^\top,
$$
and
$$
\mathrm{Var}(\hat{\beta}_{\text{ridge}} \mid X)
= \sigma^2 Q \, \mathrm{diag}\!\left(\frac{d_i}{(d_i + \lambda)^2}\right) Q^\top.
$$
For each eigenvalue $d_i > 0$ and $\lambda > 0$:
$$
\frac{d_i}{(d_i + \lambda)^2} \le \frac{1}{d_i}
\;\;\Longleftrightarrow\;\;
d_i^2 \le (d_i + \lambda)^2
\;\;\Longleftrightarrow\;\;
0 \le 2\lambda d_i + \lambda^2,
$$
which holds (strictly if $\lambda > 0$).    
Therefore, entrywise in the eigenbasis:
$$
\mathrm{diag}\!\left(\frac{d_i}{(d_i + \lambda)^2}\right)
\preceq
\mathrm{diag}\!\left(\frac{1}{d_i}\right),
$$
and by congruence with $Q$,
$$
\mathrm{Var}(\hat{\beta}_{\text{ridge}} \mid X)
\;\preceq\;
\mathrm{Var}(\hat{\beta}_{\text{OLS}} \mid X).
$$
That is, **ridge regression has smaller variance** than OLS, componentwise in the eigenbasis of $X^\top X$.

<font color="#1E90FF" size="4"><b>№7 (e)</b></font>

From the perspective of the mean squared error (MSE), could the ridge estimator have a smaller MSE than the MLE for the ordinary regression model?

From the **Mean Squared Error (MSE)** point of view, the ridge estimator can have a smaller MSE than the OLS (or MLE) estimator.    
The MSE of an estimator combines variance and bias:
$$
\mathrm{MSE}(\hat{\beta}) = \mathrm{Var}(\hat{\beta}) + [\mathrm{Bias}(\hat{\beta})]^2.
$$
OLS is **unbiased**, so:
$$
\mathrm{MSE}_{\text{OLS}}
= \mathrm{Var}(\hat{\beta}_{\text{OLS}})
= \sigma^2 (X^\top X)^{-1}.
$$
Ridge adds **bias** but reduces **variance**:
$$
\mathrm{Var}(\hat{\beta}_{\text{ridge}})
= \sigma^2 (X^\top X + \lambda I)^{-1} X^\top X (X^\top X + \lambda I)^{-1},
$$
$$
\mathrm{Bias}(\hat{\beta}_{\text{ridge}})
= -\lambda (X^\top X + \lambda I)^{-1} \beta.
$$
When predictors are **highly correlated** or the true coefficients $\beta$ are **small**,  
the **variance reduction** dominates the **bias increase**, leading to:
$$
\mathrm{MSE}_{\text{ridge}} < \mathrm{MSE}_{\text{OLS}}.
$$

<font color="#1E90FF" size="4"><b>№7 (f)</b></font>

For the above regression model, we assume that $\beta$ and $\sigma^2$ are not constant, but random variables and follow the distributions as follows: $\sigma^2$ follows an inverse Gamma distribution with parameters $\alpha_0$ and $\beta_0$, and $\beta | \sigma^2$ follows a normal distribution with mean $0_p$ and variance $\sigma^2 I_{pp}/\lambda$, where $0_p$ is a $p$-dimensional zero
vector, and $I_{pp}$ is a $p \times p$ identity matrix and $\lambda$ is a constant. Obtain the posterior distribution $p(\beta, \sigma^2 | X, Y)$.

**Model**
$$
Y \mid \beta,\sigma^2 \sim \mathcal{N}(X\beta,\;\sigma^2 I_n),\qquad
\mathbb{E}[\varepsilon\mid X]=0,\quad \mathrm{Var}(\varepsilon\mid X)=\sigma^2 I_n.
$$
**Prior**
$$
\beta \mid \sigma^2 \sim \mathcal{N}\!\left(0,\;\frac{\sigma^2}{\lambda} I_p\right)
\quad\text{(precision } \tfrac{\lambda}{\sigma^2} I_p\text{)},\qquad
\sigma^2 \sim \mathrm{Inv\text{-}Gamma}(\alpha_0,\beta_0).
$$
**Define**
$$
V_n \;=\; (X^\top X+\lambda I_p)^{-1},\qquad
\mu_n \;=\; V_n X^\top Y,\qquad
\alpha_n \;=\; \alpha_0+\frac{n}{2},
$$
$$
\beta_n \;=\; \beta_0+\tfrac{1}{2}\!\left(Y^\top Y-\mu_n^\top V_n^{-1}\mu_n\right)
\;=\; \beta_0+\tfrac{1}{2}\, Y^\top\!\bigl(I_n - X V_n X^\top\bigr)Y.
$$
**Posterior (joint)**
$$
(\beta,\sigma^2)\mid X,Y \;\sim\; \mathrm{Normal\text{-}Inverse\text{-}Gamma}\bigl(\mu_n,\;V_n,\;\alpha_n,\;\beta_n\bigr).
$$
**Posterior conditionals**
$$
\beta \mid \sigma^2,X,Y \sim \mathcal{N}(\mu_n,\;\sigma^2 V_n),\qquad
\sigma^2 \mid X,Y \sim \mathrm{Inv\text{-}Gamma}(\alpha_n,\;\beta_n).
$$

Priors
$\sigma^2 \sim \mathrm{Inv\text{-}Gamma}(\alpha_0, \beta_0)$  
$\beta \mid \sigma^2 \sim \mathcal{N}\!\left(0, \frac{\sigma^2}{\lambda} I_p\right)$  
$\lambda$ is fixed.   
Likelihood
$$
Y \mid \beta, \sigma^2 \sim \mathcal{N}(X\beta,\, \sigma^2 I_n)
$$
Posterior
$$
p(\beta, \sigma^2 \mid X, Y)
\propto (\sigma^2)^{-\frac{n+p}{2}-\alpha_0-1}
\exp\!\left[
-\frac{1}{2\sigma^2}\bigl(\|Y - X\beta\|^2 + \lambda\|\beta\|^2 + 2\beta_0\bigr)
\right]
$$
Conditional Posterior of $\beta$
$$
\beta \mid \sigma^2, X, Y
\sim \mathcal{N}(\mu_{\text{post}},\, \Sigma_{\text{post}})
$$
where  
$\mu_{\text{post}} = (X^\top X + \lambda I)^{-1} X^\top Y = \hat{\beta}_{\text{ridge}}$,  
$\Sigma_{\text{post}} = \sigma^2 (X^\top X + \lambda I)^{-1}$
Posterior of $\sigma^2$   
Let $\mathrm{SSR}(\hat{\beta}_{\text{ridge}})=\|Y - X\hat{\beta}_{\text{ridge}}\|^2$
$$
\alpha_n = \alpha_0 + \frac{n + p}{2}, \qquad
\beta_n = \beta_0 + \frac{1}{2}\bigl(\mathrm{SSR}(\hat{\beta}_{\text{ridge}}) + \lambda\|\hat{\beta}_{\text{ridge}}\|^2\bigr)
$$
Then  
$$
\sigma^2 \mid X, Y \sim \mathrm{Inv\text{-}Gamma}(\alpha_n, \beta_n)
$$
Joint Posterior
$$
\begin{aligned}
\beta \mid \sigma^2, X, Y
&\sim
\mathcal{N}\!\left(
(X^\top X + \lambda I)^{-1} X^\top Y,\;
\sigma^2 (X^\top X + \lambda I)^{-1}
\right),\\[4pt]
\sigma^2 \mid X, Y
&\sim
\mathrm{Inv\text{-}Gamma}\!\left(
\alpha_0 + \frac{n + p}{2},\;
\beta_0 + \frac{1}{2}
\bigl(\|Y - X\hat{\beta}_{\text{ridge}}\|^2 + \lambda \|\hat{\beta}_{\text{ridge}}\|^2\bigr)
\right)
\end{aligned}
$$

<font color="#1E90FF" size="4"><b>№8</b></font>

What is the general formulation of LASSO regression models? What are the differences with ridge regression? What are those situations where ridge and LASSO should be applied?

The **LASSO (Least Absolute Shrinkage and Selection Operator)** regression adds an **L1 penalty** to the standard least squares objective. For predictors $X = [x_1, x_2, \dots, x_p]$ and response $y$, the LASSO estimator $\hat{\beta}$ is defined as:
$$
\hat{\beta}_{\text{lasso}}
=
\arg\min_{\beta_0,\beta}
\left\{
\sum_{i=1}^n
\left(
y_i - \beta_0 - \sum_{j=1}^p x_{ij}\beta_j
\right)^2
+
\lambda\sum_{j=1}^p |\beta_j|
\right\}.
$$
where:
- $\lambda \ge 0$ — regularization parameter controlling shrinkage strength  
- The L1-norm penalty $\sum_{j=1}^p |\beta_j|$ encourages **sparsity** — some coefficients become exactly zero.

The **Ridge regression** estimator adds an **L2 penalty** instead of L1:

$$
\hat{\beta}_{\text{ridge}}
=
\arg\min_{\beta_0,\beta}
\left\{
\sum_{i=1}^n
\left(
y_i - \beta_0 - \sum_{j=1}^p x_{ij}\beta_j
\right)^2
+
\lambda\sum_{j=1}^p \beta_j^2
\right\}.
$$
**Key Differences**
| Feature | **LASSO (L1)** | **Ridge (L2)** |
|----------|----------------|----------------|
| **Penalty term** | $\lambda \sum_j |\beta_j|$ | $\lambda \sum_j \beta_j^2$ |
| **Shrinkage effect** | Shrinks some coefficients **exactly to zero** | Shrinks coefficients **continuously toward zero**, but never exactly zero |
| **Variable selection** | Performs automatic feature selection | Keeps all variables |
| **Computation** | Iterative (coordinate descent, LARS) | Closed form $(X^TX + \lambda I)^{-1}X^Ty$ |

**We use LASSO When:**
- Many predictors are irrelevant or redundant.  
- Goal includes **variable selection** or **model interpretability**.  
- $p > n$ (more features than samples).  
- Example: Selecting key genes in bioinformatics or key factors in finance.

**We use Ridge When:**
- Predictors are **highly correlated (multicollinearity)**.  
- All predictors are expected to contribute somewhat.  
- You need **stability** and **smooth shrinkage**, not sparsity.  
- Example: Macroeconomic forecasting or continuous risk models.
